```python
import numpy as np  # Importing NumPy library
import pandas as pd  # Importing Pandas library
import matplotlib.pyplot as plt  # Importing Matplotlib library's "pyplot" module
import seaborn as sns  # Imorting Seaborn library

import os
```

```python
data = pd.read_csv("data.csv")
```

```python
data.info()
```

```python
def diagnosis_numeric(i):
    if i=='M': return 1
    if i=='B': return 0

data['diagnosis'] = data.diagnosis.apply(diagnosis_numeric)
```

```python
data
```

```python
datanew = data.rename(columns={'Unnamed: 32': 'unnamed32'})

datanew1 = datanew.drop(['id','unnamed32'], axis=1)
                               
datanew1 = datanew1.replace(np.nan, 0)

datanew1
```

```python
X = datanew1.drop('diagnosis', axis=1)
Y = datanew1.diagnosis.values

from sklearn.model_selection import train_test_split 

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
#Logistic Regression 3 Class Classifier
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(random_state=0).fit(x_train, y_train)
y_2 = clf.predict(x_test)

accuracy_score(y_2, y_test)
```

```python
#Neural Net Classification
from sklearn.neural_network import MLPClassifier

clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(5, 2), random_state=1)

clf.fit(x_train, y_train)

y_3 = clf.predict(x_test)
accuracy_score(y_3, y_test)
```

```python
#Decision Tree Classification
from sklearn import tree
from sklearn.metrics import accuracy_score

clf = tree.DecisionTreeClassifier()
clf = clf.fit(x_train, y_train)

y_1 = clf.predict(x_test)

accuracy_score(y_1, y_test)
```
